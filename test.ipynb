```python
from snscrape.modules.twitter import TwitterSearchScraper
import snscrape
import pandas as pd
import streamlit as st
import numpy as np
```

```python
# scrape tweets
def scrape_tweets(field, n):
    tweet_list = []
    for i, tweet in enumerate(TwitterSearchScraper(field).get_items()):
        if i > n:
            break
        tweet_list.append([tweet.user.username, tweet.date, tweet.likeCount, tweet.sourceLabel, tweet.content])
    return pd.DataFrame(tweet_list, columns=['Username', 'Datetime', 'Likes', 'Source', 'Text'])
```

```python
scrape_tweets('iphone14', 10)
```
